```python
import csv
import numpy as np
from statistics import median
import numbers
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)

def get_csv_data(path, index):
    duration = []
    with open(path, "r") as csvfile:
        csv_obj = csv.reader(csvfile)
        csv_list = list(csv_obj)
        for row in csv_list[1:len(csv_list) - 2]:
            #duration.append(row[1])
            duration.append(sum([float(row[i]) for i in index]))
    return duration

path_dep_rhel = "/home/guoqingli/work/jupyter/scale-data/rhel82_4000vms_118nodes_deployments.csv"
path_dep_win = "/home/guoqingli/work/jupyter/scale-data/win10_1000vms_118nodes_deployments.csv"

path_boot_50 = "/home/guoqingli/work/jupyter/scale-data/rhel82_50vms_118nodes_bootstorm.csv"
path_boot_100 = "/home/guoqingli/work/jupyter/scale-data/rhel82_100vms_118nodes_bootstorm.csv"
path_boot_200 = "/home/guoqingli/work/jupyter/scale-data/rhel82_200vms_118nodes_bootstorm.csv"
path_boot_400 = "/home/guoqingli/work/jupyter/scale-data/rhel82_400vms_118nodes_bootstorm.csv"
path_boot_800 = "/home/guoqingli/work/jupyter/scale-data/rhel82_800vms_118nodes_bootstorm.csv"
path_boot_1600 = "/home/guoqingli/work/jupyter/scale-data/rhel82_1600vms_118nodes_bootstorm.csv"
path_migration_1500 = "/home/guoqingli/work/jupyter/scale-data/4970_vms_zone-2_1500vms_migration.csv"

dep_paths = [path_dep_rhel, path_dep_win]
boot_paths = [path_boot_50, path_boot_100, path_boot_200, path_boot_400, path_boot_800, path_boot_1600]
migration_paths = [path_migration_1500]

def get_data_from_multi_path(paths, index):
    res = []
    for path in paths:
        res.append(get_csv_data(path, index))
    return res
                             
def time_series_plot(ydata, labels, title, xlabel, ylabel, xdata=None):
    if xdata is None:
        for i in range(len(ydata)):
            plt.plot(ydata[i], label=labels[i])
        plt.xticks([])
    else:
        for i in range(len(ydata)):
            plt.plot(xdata, ydata[i], label=labels[i])
        plt.grid()
    plt.xlabel(xlabel)
    # plt.xticks(np.arange(min(xdata),max(xdata), 200))
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.17),fancybox=True, shadow=True, ncol=len(ydata))
    plt.ylim(bottom=0)
    plt.tight_layout()
    # plt.savefig('./plots/'+ title + '.png')
    plt.show()

def box_plot(data, title, xlabel, labels, ylabel, vert):
    fig, ax = plt.subplots()
    #colors = ['#D2B4DE', '#F7DC6F']
    flierprops = dict(marker='o', markerfacecolor='r', markersize=2, linestyle='none', markeredgecolor='g')
    # data = [[np.log2(n) for n in k] for k in data ]
    bp = ax.boxplot(data, labels=labels, patch_artist = True, flierprops=flierprops, vert = vert, widths = 0.2, meanline=True, showmeans=True, showcaps=True, showfliers=True, manage_ticks=True)
    
   # for patch, color in zip(bp['boxes'], colors): patch.set_facecolor(color)
    
    for median in bp['medians']:
        median.set(color ='red', linewidth = 1)
    
    for median in bp['means']:
        median.set(color ='blue', linewidth = 1)
    
    ax.set_title(title)
    ax.set_xlabel(xlabel, fontsize=10)
    ax.set_ylabel(ylabel, fontsize=10)
    ax.grid(axis="y", which='major', color='#CCCCCC', linestyle='--', linewidth=0.5)
    # ax.set_yscale('log', base=2)
    # for m in bp['means']:
    #     print(m.get_ydata()[0])
    
    for i, line in enumerate(bp['means']):
        x, y = line.get_xydata()[1]
        if vert == 1:
            text = ' μ={:.2f}'.format(y)
        else:
            text = ' μ={:.2f}'.format(x)
        ax.annotate(text, xy=(x, y+0.02), fontsize=6)
    
    for i, line in enumerate(bp['medians']):
        x, y = line.get_xydata()[1]
        if vert == 1:
            text = ' x͂={:.2f}'.format(y)
        else:
            text = ' x͂={:.2f}'.format(x)
            ax.annotate(text, xy=(x-4, y-0.25), fontsize=6)
        
    for i, line in enumerate(bp['whiskers']):
        x, y = line.get_xydata()[1]
        if vert == 1:
            text = ' {:.2f}'.format(y)
        else:
            text = ' {:.2f}'.format(x)

        if vert ==0:
            ax.annotate(text, xy=(x-7, y+0.06), fontsize=6)
        else:
            if i % 2 == 0:
                ax.annotate(text, xy=(x-0.1, y - 35), fontsize=6)
                # ax.annotate(text, xy=(x-0.1, y - 3.5), fontsize=6)
            else:
                ax.annotate(text, xy=(x-0.1, y + 10), fontsize=6)
                #ax.annotate(text, xy=(x-0.1, y+5), fontsize=6)
    # plt.gcf()
    if vert == 0:
        fig.set_size_inches(6, 5)
    else:
        fig.set_size_inches(5, 7)
    plt.tight_layout()
    #plt.ylim(bottom=0)
    plt.savefig('/home/guoqingli/work/jupyter/scale-plot/{}'.format(title), dpi=1200)
    plt.gcf().set_dpi(300)
    plt.show()

dep_data = get_data_from_multi_path(dep_paths, [2])
boot_data = get_data_from_multi_path(boot_paths,  [2])
# migration_data = get_data_from_multi_path(migration_paths, [1])
print(boot_data)
box_plot([boot_data[0], labels=['50', '100', '200', '400', '800', '1600'], title="RHEL VM Boot Storm", ylabel="Seconds", xlabel="No. of VMs", vert=1)
box_plot(dep_data, labels=['RHEL', 'Windows'], title="VM Deployment", xlabel="Seconds", ylabel="", vert=0)

# d ={}

# old_nodes = migration_data[0]
# for node in old_nodes:
#     if node in d:
#         d[node] += 1
#     else:
#         d[node] = 1

# print(d)


```
