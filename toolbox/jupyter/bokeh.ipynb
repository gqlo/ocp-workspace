```python
from bokeh.plotting import figure, show, output_notebook, output_file
from bokeh.models import ColumnDataSource, CustomJS, Select
from bokeh.layouts import column
from bokeh.io import curdoc
import json

def create_interactive_plot(datasets):
    output_notebook()
    blk_size = list(datasets.keys())

    # Create the ColumnDataSource
    source = ColumnDataSource(data=datasets[blk_size[0]])

    # Create the figure
    p = figure(title="Switching between plots")
    p.line('x', 'y', source=source)

    # Create the CustomJS callback arguments dynamically
    args = {'source': source}
    for blk in datasets:
        args[blk] = datasets[blk]

    # Convert datasets to JSON string for JavaScript usage
    data_sets_json = json.dumps(datasets)
    js_code = f"""
        var data_sets = {data_sets_json};
        console.log(data_sets);
        console.log(cb_obj.value);
        source.data = data_sets[cb_obj.value];
        source.change.emit();
    """

    print("js_code=", js_code)
    # Create the CustomJS callback
    callback = CustomJS(args=args, code=js_code)

    # Create a Select widget with options for all datasets
    select = Select(title="Select Dataset", value=blk_size[0], options=blk_size)
    select.js_on_change('value', callback)

    # Layout and add to the current document
    layout = column(select, p)
    curdoc().add_root(layout)

    # Output to an HTML file
    output_file("bokeh_plot.html")
    show(layout)

# Define multiple data sets
datasets = {
    "4k": {'x': [1, 2, 3, 4, 5], 'y': [6, 7, 2, 4, 5]},
    "8k": {'x': [1, 2, 3, 4, 5], 'y': [5, 4, 3, 2, 1]},
    "16k": {'x': [1, 2, 3, 4, 5], 'y': [8, 5, 3, 2, 7]},
    "32k": {'x': [1, 2, 3, 4, 5], 'y': [1, 3, 6, 8, 5]},
}

create_interactive_plot(datasets)
```
